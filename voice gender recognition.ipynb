```python
#seperating features and labels
x=data.iloc[:,:-1]
y=data.pop('label')
```

```python
x.head()
```

```python
#encoding the labels to 1--> male and female-->0
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
y
```

```python
#scale the data between -1 and 1
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
#splitting the dataset into train test
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=30)
```

```python
#running svm with default hyperparameter
from sklearn.svm import SVC
svc=SVC()
svc.fit(xtrain,ytrain)
```

```python
support_vectors=svc.support_vectors_
pd.DataFrame(support_vectors,columns=data.columns).head()
```

```python
svc.n_support_ #no. of support vectors
```

```python
svc.support_ #indices of support vectors
```

```python
from sklearn import metrics
ypred=svc.predict(xtest)
metrics.accuracy_score(ytest,ypred)
```

```python
%matplotlib inline
cmat=metrics.confusion_matrix(ytest,ypred)
import seaborn as sn

sn.heatmap(cmat,annot=True)
```

```python
svc
```

```python
#svm with linear kernel
svc=SVC(kernel='linear')
svc.fit(xtrain,ytrain)
```

```python
ypred=svc.predict(xtest)
metrics.accuracy_score(ytest,ypred)

```

```python
cmat=metrics.confusion_matrix(ytest,ypred)
sn.heatmap(cmat,annot=True)
```

```python
#polynomial kernel
sv=SVC(kernel='polynomial')
svc.fit(xtrain,ytrain)
ypred=svc.predict(xtest)
metrics.accuracy_score(ytest,ypred)
```

```python
#k fold cross validation
from sklearn.model_selection import cross_val_score
```

```python
svc=SVC(kernel='linear')
scores=cross_val_score(svc,x,y,cv=10,scoring='accuracy')
print(scores)
```

```python
scores.mean()
```

```python
svc=SVC(kernel='rbf')
scores=cross_val_score(svc,x,y,cv=10,scoring='accuracy')
print(scores)
```
